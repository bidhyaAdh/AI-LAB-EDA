```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

```python

# Load dataset
def load_dataset(file_path):
  return pd.read_csv(file_path)


df = load_dataset("/content/drive/MyDrive/AI EDA lab/adult.csv")
df.head(5)
```

```python
sns.countplot(data=df, x='gender')
df['workclass'].value_counts()
```

```python

df['gender'].value_counts().plot(kind='pie',autopct='%0.2f')
```

```python

# Histogram treat as categorical data distribution of data

plt.hist(df['age'])
```

```python
sns.distplot(df['age'])
```

```python
sns.boxplot(df['age'])
```

```python
# Scatter plot(Numerical - Numerical) for the relation between two column
sns.scatterplot(x='relationship', y='age',hue='race', style='workclass',size='occupation' ,data=df)
plt.show()
```

```python
sns.barplot(x='marital-status',y='age',hue='gender',data=df)
```

```python
sns.boxplot(x='gender',y='age',hue='marital-status',data=df)
```
